3.2.1.1 **Missing Data Information**

For missing data treatment in the dataset, both categorical and non-categorical variables are initially encoded from string format. Following this, the SimpleImputer method is applied with the strategy 'most_frequent' to all fields containing null values. This imputation technique ensures that missing values in each column are replaced with the most frequently occurring value within that column. After the imputation process, no columns in the dataset contain null values, ensuring completeness of the data for further analysis.

3.2.1.2 **Missing Data Exclusion [MCAR | MAR (> 50%)]**
In case any null values persist after the initial imputation process, the corresponding records containing those null values are removed from the dataset. If null values persist in specific variables even after removing the records, those particular variables are then considered for elimination from the dataset. This iterative approach ensures the integrity of the dataset by systematically handling missing data at both the record and variable levels.

3.3 **Numeric Encoding of data:**

 Numerical encoding of categorical data : Following the initial missing data treatment, numeric encoding is performed using the ordinal encoder. This encoding technique assigns numerical labels to categorical variables, preserving the ordinal relationships within each variable. This step facilitates the transformation of categorical data into a format suitable for analysis and modeling, ensuring compatibility with various machine learning algorithms and statistical techniques.

By utilizing the median and interquartile range (IQR) instead of mean and standard deviation, the Robust Scaler ensures that the presence of outliers does not unduly influence the scaling process. This approach helps to maintain the integrity of the data while effectively normalizing the features for downstream analysis.

```python
subset = df_ppd[['so2_robnorm','no2_robnorm']]
# Plotting Scatter Plot for NO2 vs SO2 with spectral colormap
plt.scatter(subset['no2_robnorm'], subset['so2_robnorm'], c=subset['no2_robnorm'], cmap='Spectral')
plt.xlabel('NO2 Concentration')
plt.ylabel('SO2 Concentration')
plt.title('NO2 vs SO2 Concentration')
plt.colorbar()
plt.grid()
plt.show()
```

```python
wcssd = [] # Within-Cluster-Sum-Squared-Distance
nr_clus = range(1,11) # Number of Clusters
for k in nr_clus:
    kmeans = kmclus(n_clusters=k, init='random', random_state=111)
    kmeans.fit(df_AirQualityIndex)
    wcssd.append(kmeans.inertia_)
plt.plot(nr_clus, wcssd, marker='x')
plt.xlabel('Values of K')
plt.ylabel('Within Cluster Sum Squared Distance')
plt.title('Elbow Curve for Optimal K')
plt.show()
```

```python
km_2cluster = kmclus(n_clusters=2, init='random', random_state=222)
km_2cluster_model = km_2cluster.fit_predict(subset); km_2cluster_model
```
